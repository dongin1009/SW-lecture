## 스캐터 플롯 (scatterplot)

### 개요
- 두 변수에 대해 상대적인 관계를 나타내는 가장 기본적인 방법
- 점들이 여기저기 흩어진 모양의 그래프를 의미함
- seaborn.scatterplot(), pandas.lmplot()으로도 가능
- https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
import seaborn as sns
import pandas as pd
df = pd.read_csv("./data/iris.csv")
sns.scatterplot(x='petal.length', y='petal.width', data=df)
```

- hue : 변수의 값에 따라 점들을 그룹화하고 색으로 구분

```python
sns.scatterplot(x='petal.length', y='petal.width', hue='species', data=df)
```

- style : 변수의 값에 따라 점들을 그룹화하고 모양으로 구분

```python
sns.scatterplot(x='petal.length', y='petal.width', hue='species', style='species', data=df)
```

- lmplot을 이용(fit_reg : 회귀 선)

```python
sns.lmplot(x='petal.length', y='petal.width', hue='species', data=df, fit_reg=False, palette='bright', markers=['o','x','v'])
```

```python
sns.lmplot(x='petal.length', y='petal.width', hue='species', data=df, fit_reg=True, palette='bright', markers=['o','x','v'])
```

## 히스토그램 (histogram)

- 도수분포표를 그래프로 나타낸 것
- 어떠한 변수에 대해서 구간별 빈도수(즉, 분포)를 나타낸 그래프

```python
sns.distplot(df['petal.length'], bins=15)
```

## 조인트 플롯(jointplto)
### 개요
- 스캐터 플롯과 히스토그램을 두 변수에 대해서 결합
- http://seaborn.pydata.org/generated/seaborn.jointplot.html

- 가장 기본적인 jointplot, 기본이 scatter, marginal_kws : histogram의 bin 개수 설정

- kde=true, 커널밀도추정 기능 설정

## 커널 밀도 추정

### 밀도 측정
- 변수의 sample space : 변수에 대해 측정 가능한 모든 값의 범위
- 데이터 : 실제 측정된(관측된, 수집된) 값, sample space의 일부의 값
- 데이터를 이용해서 sample space 전체에 대한 밀도를 추정하는 것

### 커널 밀도 추정 (Kernel Density Estimation, KDE)
- 밀도 추정시 커널 함수를 이용하는 밀도 추정 방법
- 커널 함수 : 원점을 중심으로 대칭이며 적분 값이 1인 양의 함수
- 커널함수와 결합하며 밀도추정 수행

```python
sns.jointplot(x='petal.length', y='petal.width', data=df, kind='scatter', marginal_kws=dict(bins=10))
```

```python
sns.jointplot(x='petal.length', y='petal.width', data=df, kind='kde')
```

## 페어 플롯(Pair Plot)
### 개요
- 복수의 변수들에 대해서 모든 두 변수간의 스캐터 플롯을 출력
- 대각선에는 단일 변수에 대한 히스토그램으로 보여줌

```python
from sklearn import datasets

boston_data = datasets.load_boston()
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)

sns.pairplot(data=df)
```

- vars 파라미터 : pairplot을 수행할 변수만 지정

```python
vars_to_plot = ['CRIM', 'AGE', 'DIS']
sns.pairplot(data=df, vars=vars_to_plot)
```

## 입력 데이터 클리닝
### 개요
- 실제 데이터 집합은 지저분(dirty)하며 유용한 인사이트를 얻기 위해서는 데이터의 무결성이 보장되어야 함
- 결측값 및 손상된 값은 정확한 데이터 분석을 어렵게 함
- 결측값 외에도 속성 스케일링, 변수 유형(연속, 불연속) 등이 분석에 영향을 미침
- 입력 데이터 클리닝은 결측값 처리, 정규화와 표준화 등을 수행하여 분석의 정확도를 높이는 것

## 결측값 처리
### 개요
- 결측값은 데이터 마이닝을 망치는 주범
- 한 열, 한 행이 없을 수도 있고, 특정 레코드의 특정 변수 값이 없을 수도 있음
- 결측값을 찾고 상황에 따라 제거 및 대체하는 과정이 필수

### 결측값 찾기
- pandas를 이용해서 데이터를 로딩하면 결측값은 NaN(Not a Number)으로 표시됨

```python
df = pd.read_csv("./data/iris_missing_values.csv")
df.index.name='record'
print(df.head())
```

- pandas의 isnull() 기능 이용
- isnull()은 데이터 프레임 전체, 특정 행, 특정 열에 대해서 동작할 수 있음

```python
# 모든 레코드(행)에 대해서 null값의 여부를 true/false로 반환
print(df['sepal.length'].isnull())
# 하나라도 NaN이 있으면 true
print(df['sepal.length'].isnull().values.any())
# NaN의 개수 반환
print(df['sepal.length'].isnull().values.sum())
```

### 결측값 대체

- 어떤 값으로 대체할 것인가?? -> 적절한 imputing 기법 이용(평균, 중앙값, 최빈값, 상수 등)

```python
df_example = df['sepal.length'].fillna('example')
print(df_example.head())
```

### 결측값 제거

- axis = 0 : 행
- axis = 1 : 열
